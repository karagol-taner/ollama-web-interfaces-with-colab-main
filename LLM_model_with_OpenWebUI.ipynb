```python

# Open Ngrok tunnel to Ollama server
ollama_tunnel = ngrok.connect("8082", "http")
public_url = ollama_tunnel.public_url
print(f" * ngrok tunnel \"{public_url}\" -> \"localhost:8082\"")
```
